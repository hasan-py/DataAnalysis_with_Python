```python
# -----------BroadCasting Operations------------

newnba = pd.read_csv("nba.csv")
newnba.head()
```

```python
# We can use this column and row for calculations

newnba["Salary"].add(5000)
newnba["Salary"]-500

newnba["Age"]
```

```python
# ------------Drop and Null Value----(Please Not Use it)-------------
newnba.head()
```

```python
newnba.dropna(how="all") # Here we Delete all null row by this method
newnba.dropna(how="any") # Here any delete all of null row and columns

newnba.dropna(how="all" ,inplace=True) # For replace 
newnba.dropna(axis=1) # all row remove


```

```python
# ----------------fillna method-------(*****)---------

newnba1 = pd.read_csv("nba.csv")
newnba1.head()

newnba1.fillna(0).head() # it will replace all null by 0
newnba1["Salary"].fillna(0,inplace=True) # here we select a particulaer columns
newnba1["College"].fillna("No College", inplace=True)
newnba1.head() # It just change college row null
```

```python
# --------------sort_values()-----------------
newnba2 = pd.read_csv("nba.csv")
newnba2.head()

newnba2.sort_values("Name") # HEre we sort Name column
newnba2.sort_values("Name", ascending=False) # Z-A 

newnba2.sort_values("Age",ascending=False)

newnba2.sort_values("Salary") # Here na position mean null
newnba2.sort_values("Salary",na_position="first")
```

```python
# ---------sorting Multiple Columns----------
newnba2.head()
```

```python
newnba2.sort_values(["Team","Name"],ascending=False) #Here We sort multiple columns
newnba2.sort_values(["Team","Name"],ascending=[False,True],inplace=True) # two diffrent type sorting in 1 line
newnba2.head() # inplace=True must
```
