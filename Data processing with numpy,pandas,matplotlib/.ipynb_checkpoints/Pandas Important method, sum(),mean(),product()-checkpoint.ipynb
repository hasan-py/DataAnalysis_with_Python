```python
newdf = pd.DataFrame(np.random.randint(1,334,(334,6)))
```

```python
newdf.head()
```

```python
newdf.describe() #------Here describe function give count,max and etc item---------
newdf.dtypes #check Datatype
newdf.index #Check Index
newdf.columns #check columns

newdf.to_numpy() #To create numpy array 

newdf.T #for traverse 5 row 334 columns

newdf.columns = list("ABCDE") #Here We Can Change our Columns Name

newdf.head()
newdf2 = newdf.copy() #For copy
newdf2.loc[0]['A'] = 100000 #Here loc method change our value


# -------axis=1 means columns and axis=0 means row----------
newdf = newdf.drop('F',axis=1) #Here drop Function use for Delet Colum and row



```

```python
# --------For show dataframe item and value----------
newdf3 = pd.DataFrame(np.random.randint(1,334,(334,6)))
newdf3.columns = list("ABCDEF")

# -------------loc-----------
newdf3.loc[[1,2],["C","D"]] #Here loc function Shows the exact item or value
newdf3.loc[[1],["C"]] = 3333  #Here we change it
newdf3.loc[[1,2],["C","D"]] #it not change ... 
newdf3.loc[[1,2],:] #This is show 1,2 row
newdf3.loc[:,["A","B"]] #This is show A,B columns



newdf3.head(2)
newdf3.iloc[0,0] #This is for iloc


newdf3.loc[[0,2],["A","B","C"]] # This is very helpful
```

```python
# -----------Droping or Deleting DataFrame row and columns-----

newdf3.drop([0,1], axis=0) #it delete the row 1,2
# newdf3.drop(["D","F"], axis=1) #it Delete columns A,B by axis=1
# newdf3 = newdf3.drop(["D","F"], axis=1) #HEre it delete those item
newdf3.head(10)
# newdf3.drop(["C"],axis=1,inplace=True) #It Delete The Item
newdf3.reset_index(drop=True,inplace=True) # It make True

# ---------Always use loc ---------------
newdf3.head()
newdf3.loc[[1]]
```
