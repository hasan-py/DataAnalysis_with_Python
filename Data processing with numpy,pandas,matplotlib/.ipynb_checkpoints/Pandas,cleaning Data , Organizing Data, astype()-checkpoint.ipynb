```python
import pandas as pd
```

```python
# -------------Rank Method------------------
nba = pd.read_csv("nba.csv").dropna(how="all")
nba["Salary"] = nba["Salary"].fillna(0).astype(int) # here astype makes all value int
nba.head()


```

```python
# -------------Implement our rank method----------

nba["Salary"].rank(ascending=False).astype(int)
nba["SalaryRank"] = nba["Salary"].rank(ascending=False).astype(int)
nba.sort_values(by = "Salary", ascending=False, na_position="last",inplace=True)
nba.head()
```

```python
# ----------------------------------------------- Filtering/Cleaning DataFrame ------------------------------------------------

emp = pd.read_csv("employees.csv")
emp.head()
```

```python
emp.info()
```

```python
emp["Start Date"].head()
# ---------Here we convert date to a pandas datetime field 
pd.to_datetime(emp["Start Date"])

emp["Start Date"] = pd.to_datetime(emp["Start Date"])
emp.sort_values(by = "Start Date",inplace=True)
emp
```

```python
# ------Convert last login-------------
emp["Last Login Time"].head()

emp["Last Login Time"] = pd.to_datetime(emp["Last Login Time"])
emp.head()
```

```python
# --------------Senior management -------------
emp["Senior Management"].astype("bool")
emp["Senior Management"] = emp["Senior Management"].astype("bool")

# ---------------Gender Change-----------------
emp["Gender"].astype("category")
emp["Gender"] = emp["Gender"].astype("category")
emp.info()
```

```python
# ------------Anouther Cleaning system----(*****)------------

df = pd.read_csv("employees.csv",parse_dates=["Start Date","Last Login Time"])
df["Senior Management"] = emp["Senior Management"].astype("bool")
df["Gender"] = emp["Gender"].astype("category")
df.info()
df
```
