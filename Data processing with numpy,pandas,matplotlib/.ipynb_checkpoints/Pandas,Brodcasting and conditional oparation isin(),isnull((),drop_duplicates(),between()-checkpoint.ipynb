```python
df = pd.read_csv("employees.csv")
df.info()
```

```python
# -----------Convert unclean massive data to clean data ----------------

# ---Convert string to DateTime ----
pd.to_datetime(df["Start Date"])
df["Start Date"] = pd.to_datetime(df["Start Date"])
df["Last Login Time"]

df["Last Login Time"] = pd.to_datetime(df["Last Login Time"])

df["Last Login Time"]
```

```python
# ---- Convert String to Boolean -----
df["Senior Management"] = df["Senior Management"].astype("bool")
df["Senior Management"]

# ------convert some informaion by Cetagory-----
df["Gender"] = df["Gender"].astype("category")
df.info()
```

```python
# -----------------------Filtering Data with Condision -------------------

mask = df[df["Team"] == "Finance"]
mask = mask[mask["Gender"] == "Male"]
mask = mask[mask["Salary"] < 80000 ]
mask = mask[mask["Senior Management"] == True]
mask
```

```python
df["Bonus %"] < 1.5
df["Bonus %"] > 1.5

df["Start Date"] <= "1985-01-01"
mask["Start Date"] >= "2000-01-01"

mask[mask["Start Date"] >= "2000-01-01"]
```

```python
# ------Filter Dataframe with AND & OR Operators| 2018 |Python 3---------

```

```python
# -------------AND Operator -------------------
mask1 = df["Gender"] == "Male"
mask2 = df["Team"] == "Marketing"
df[mask1 & mask2]
```

```python
# ---------------OR Operator------------------

mask3 = df["Senior Management"]
mask4 = df["Start Date"] < "1990-01-01"
df[mask3 | mask4]
```

```python
mask5 = df["First Name"] == "Robert" 
mask6 = df["Team"] == "Client Services"
mask7 = df["Start Date"] > "2016-06-01"
df[(mask5 & mask6) | mask7]
```

```python
# ---------------------.isin() Method in Pandas-----------------------
# This is for providing multiple argument
```

```python
ma = df["Team"].isin(["Legal","Sales","Product"])
ma1 = df["Senior Management"].isin([True,False])

df[(ma & ma1)]
```

```python
# -------------isnull() & .notnull()--------------
df.head()
```

```python
nu = df["Team"].isnull() #This is give all null vakue
notnu = df["Gender"].notnull() # this is give not null value 
df[notnu]
```

```python
# -------.between() Method in Pandas| 2018 |Python 3--------
```

```python
df.head(4)
df[df["Salary"].between(60000,70000)]
df[df["Bonus %"].between(2.0,5.0)]

# -----some filtering with data ------
teams = df["Team"].isin(["Legal"])
males = df["Gender"].isin(["Male"])
logintimes = df["Last Login Time"].between("10:00AM","11:00AM")
df[(males & logintimes) & teams]
```

```python
# ------- .duplicated() Method --------------

df.sort_values(["First Name","Team"], inplace=True)

df["First Name"].duplicated()

df[df["First Name"].duplicated()].head()

df["First Name"].duplicated(keep = 'last')
df[df["First Name"].duplicated(keep = 'last')].head()

df["First Name"].duplicated(keep = False)
df[df["First Name"].duplicated(keep = False)].head()

# --------- (~) Shift+ESC button for reverse oparation --------- 
maskk = ~df["First Name"].duplicated(keep = False)
df[maskk]
```

```python
# --------------- Drop Duplicate ------------

len(df.drop_duplicates(subset=["First Name"],keep="first"))
df.drop_duplicates(subset=["First Name"],keep="first")

df.drop_duplicates(subset=["First Name","Team"],keep="last") #Here we use or condistions
```
