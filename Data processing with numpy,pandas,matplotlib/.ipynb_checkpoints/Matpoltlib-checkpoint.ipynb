```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6]
y = [2,3,4,6,7,3]

plt.plot(x,y) #This is for setup plot
plt.xlabel('x-axis') #this is for set x axis name
plt.ylabel('y-axis') #This is for set y axis name
plt.title('Our First Plot') #This is for set Title name for the plot
plt.show() # This command for show the graph
```
