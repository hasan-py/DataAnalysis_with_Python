```python
import pandas as pd

bond = pd.read_csv("jamesbond.csv", index_col="Film")
bond.sort_index()
bond.head()
```

```python
# ------------------- Changing Value in specific Cell or row ------------

# Single value Change
bond.ix["Dr. No"]
bond.ix["Dr. No","Actor"] = "Sir Sean Connery"
bond.ix["Dr. No"]

```

```python
# Multiple value Change

bond.ix["Dr. No",["Box Office","Budget","Bond Actor Salary"]]
bond.ix["Dr. No",["Box Office","Budget","Bond Actor Salary"]]= [555,44,8]
bond.ix["Dr. No",["Box Office","Budget","Bond Actor Salary"]]
```

```python

# So we need ix[] method to change orginal dataframe

mask = bond["Actor"] == "Sean Connery"
bond[mask]
# This is reguler Method
bond[mask]["Actor"] = "Sir Sean Connery"
```

```python
# ------ Using ix[] method and Change orginal DataFrame

bond.ix[mask,"Actor"] = "Sir Sean Connery"
# We can use mask var for bond["Actor"]=="Roger Moore"
bond[bond["Actor"] == "Sir Roger Moore"]
roger = bond["Actor"]=="Sir Roger Moore" # Here Come all Details about Roger Moore
bond.ix[roger,"Actor"]  #Here Comes Just Actor name 
bond.ix[roger,"Actor"] = "Roger Moore" #Now Change the name
bond[bond["Actor"] == "Roger Moore"] #Now we see in orginal dataframe change
```

```python
mask1 = bond["Actor"] == "Pierce Brosnan"
bond.ix[mask1,"Actor"] = "S Pierce Brosnan"
bond[bond["Actor"] == "S Pierce Brosnan"] # Now we see the changeing Data Frame with actor Name
```

```python
# -------------Rename Index Labels or Columns in Pandas------------

# Here We use Dict

# -----For Columns-------
bond.rename(columns={"Year":"Release Date","Box Office":"Reveneu"},inplace=True)
bond.head()
# -----For Index--------
bond.rename(index={"Dr. No":"Doctor No","GoldenEye":"Golden Eye"},inplace=True)
bond
```

```python
# ----------------------------------- Query()--------------------------------------
import pandas as pd

bond = pd.read_csv("jamesbond.csv", index_col="Film")
bond.sort_index()
bond.head()
```

```python
# ----- Here We Change our Columns name " " to "_" for query ------------

bond.columns = [columns_name.replace(" ","_") for columns_name in bond.columns]
bond.columns
```

```python
# Query with conditions
bond.query(" Actor == 'Sean Connery' ")
bond.query(" Year >= 2000 ")
bond.query(" Box_Office >= 900 ")
bond.query(" Actor != 'Roger Moore' ")
bond.query(" Budget <= 100 ")
bond.query(" Budget <= 20 ")

# Two condition Both
bond.query(" Box_Office <= 500 and Actor == 'Roger Moore' ")
bond.query(" Actor in ['Timothy Dalton','George Lazenby'] ")
bond.query(" Actor not in ['Timothy Dalton','George Lazenby'] ")

bond.query(" Box_Office >= 500 and Actor == 'Sean Connery' ")
bond.query(" Box_Office >= 500 and Actor == 'Roger Moore'  ")

bond.query(" Box_Office >= 500 and Actor == ['Sean Connery','Roger Moore'] ")
```

```python
# sorting Value Then Query
bond.sort_values("Year",inplace=True)
bond[bond["Actor"] == "Sean Connery"]
bond.query(" Box_Office >= 500 and Actor == 'Sean Connery' ")
bond.query(" Year <= 1980 and Actor == ['Sean Connery','Roger Moore']")
bond.query(" Year <= 1970 and Actor == 'Sean Connery' and Box_Office >= 500 ")

bond.query(" Box_Office >= 500 ")
```

```python
# --------------- Delete/Drop()---------------------

bond.drop("A View to a Kill")
bond.drop("Casino Royale")
```

```python
bond.drop(["GoldenEye","Die Another Day","Dr. No"]) # Multiple Delete
# Deleteing Columns .. 0 for row , 1 for columns
bond.drop(['Box_Office','Actor'],axis=1)

# ------------------ This is perfect Droping -----------------
bond.drop(columns=["Box_Office","Year","Actor"],index=["Dr. No","Moonraker"])
 
bond.pop("Actor") # this pop method makes saparate the Actor columns
bond
```
