```python
import pandas as pd

nba = pd.read_csv("nba.csv")
nba.head()
```

```python
# ------------------Brodcasting operator------------------
nba["Salary"]+5
nba["Age"]-5
nba["Weight"]*2
nba["Weight"]/5
# ------This is for replace--------
nba["Weight in KG"] = nba["Weight"]*2
nba
```

```python
# This is for drop columns
nba.drop(columns=["Weight in KG"], axis=1,inplace=True)
```

```python
# Drop Rows with Null values 

nba.dropna() #drop all thing
nba.dropna(how="any")
nba.dropna(how="all")
```
