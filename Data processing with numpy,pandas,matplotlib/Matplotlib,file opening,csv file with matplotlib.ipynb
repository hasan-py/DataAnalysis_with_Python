```python
import matplotlib.pyplot as plt
import csv
import numpy as np

x = []
y = []

with open('test.txt','r') as csvfile: #Here for txt file
    plots = csv.reader(csvfile)
    
    for col in plots:
        x.append(col[0])
        y.append(col[1])

plt.plot(x,y,label='Text File')
plt.xlabel('x-axis')
plt.ylabel('y-label')

plt.title('Test Graph')
plt.legend()

plt.show()

```
