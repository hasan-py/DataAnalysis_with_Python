```python
#----------------------------Line plot---------------------------- 
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(0.0,2.0,0.01)
s = 1 + np.cos(2*np.pi*t) #np.pi is pi values

plt.plot(t,s,'g') #here g indigate Color '*', '--' indigate * and -
plt.grid() # Here grid help to create grid

plt.xlabel('Time(t)')
plt.ylabel('Voltage(mV)')

plt.title('Cosine Wave Plot(cos(x))')
plt.show()
```

```python
# ------------------------Subplot------------------------

x1 = np.linspace(0.0,5.0)
x2 = np.linspace(0.0,2.0)

y1 = np.cos(2*np.pi*x1)*np.exp(-x1)
y2 = np.cos(2*np.pi*x2)

plt.subplot(2,1,1)
plt.plot(x1,y1,'o-')
plt.title('Subplot-1')
plt.xlabel('x1')
plt.ylabel('amp(y1)')
#Subplot
plt.subplot(2,1,2)
plt.plot(x2,y2,'.-')
plt.title('Subplot-2')
plt.xlabel('X2')
plt.ylabel('amp(y2)')

plt.show()
```

```python
# ----------------------------BarPlot----------------------------

x = [1,2,3,4,5]
y = [10,23,36,40,5]

tick_label = ['One','Two','three','Four','five']
plt.bar(x,y,tick_label=tick_label,width=0.7,color=['green','blue','red','yellow','orange'])
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Bar Graph')
plt.show()
```

```python
# -------------------------Histogram----------------------------
ages = [2,50,70,40,45,60,34,44,7,13,51,19,17,81,21,22,19,20]
range = (0,100)
bins = 10

plt.hist(ages,bins,range,color='green',histtype='bar',rwidth=0.7)
plt.xlabel('ages')
plt.ylabel('Bins')
plt.title('histogram')
plt.show()
```

```python
# -----------------------Scatter Plot---------------------
x = np.arange(1,11)

y = np.random.randint(0,10,10)


plt.scatter(x,y,label='Stars',color='green',marker='*',s=30)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')

plt.legend() #this is for label show
plt.show()
```

```python
# -------------------------Pie Chart----------------------

activites = ['eat','sleep','play','work']
slices = [3,7,8,6]
color = ['r','g','m','b']

plt.pie(slices,labels=activites,colors=color,startangle=90,shadow=True,explode=(0,0.2,0,0),autopct='%1.2f%%')
plt.legend()
plt.show()
```

```python
# -----------------------Mergeing Plot------------------------

x = np.arange(0,2*np.pi,0.1)

y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x,y1,label='Sin')
plt.plot(x,y2,label='Cos')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.legend()
plt.title('Sine cos Function')
plt.show()








```
