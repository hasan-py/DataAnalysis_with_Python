```python
# -------Math Function in Pandas------------

google = pd.read_csv("google_stock_price.csv",squeeze=True)

google.head()
```

```python
# ------------count-----------
google.count()
len(google)
```

```python
google.sum()

google.mean() #Avarage
google.sum()/google.count() #Avarage

google.min()
google.max()

google.std()

google.median() #This is for center Value
```

```python
# Index min and min
google.idxmax()
google.idxmin() 
```

```python
google.head()
```

```python
# --------------Apply Function ---------------

def classify_per(number):
    if number<300:
        return "ok"
    elif number >= 300 and number <650:
        return "Satisfactory"
    else :
        return "Inredible"

google.apply(classify_per).tail(5)
google.apply(lambda stock_price: stock_price+1)
```
